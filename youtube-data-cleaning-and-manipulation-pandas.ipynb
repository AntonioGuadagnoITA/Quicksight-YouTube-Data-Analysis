```python
# Importing Libraries
import pandas as pd
from datetime import datetime
```

```python
# Reading the df
df = pd.read_csv('/kaggle/input/youtube-analytics/Youtube Analytics.csv')
print(df.columns)
df.head()
```

```python
# Converting "Ora pubblicazione video"'s format to "YYYY-MM-DD"
df['Ora pubblicazione video'] = pd.to_datetime(df['Ora pubblicazione video'], format='%b %d, %Y').dt.strftime('%Y-%m-%d')

# Ordering the df by "Ora pubblicazione video" descending
df = df.sort_values(by='Ora pubblicazione video', ascending=False)
df = df.reset_index(drop=True)
df.head()
```

```python
# Dropping useless columns
df = df.drop(columns='Contenuti')
df = df.drop(columns='Tempo di visualizzazione (ore)')
```

```python
# Function that converts the video duration in seconds
def converti_tempo(tempo_str):
    h, m, s = tempo_str.split(':')
    return int(h) * 3600 + int(m) * 60 + int(s)

# Function that converts 'Percentuale media visualizzata (%)' into a float 
def converti_percentuale(percentuale_str):
    return float(percentuale_str.replace(',', '.'))

# Applying transformations
df['Durata di visualizzazione media'] = df['Durata di visualizzazione media'].apply(converti_tempo)
df['Percentuale media visualizzata (%)'] = df['Percentuale media visualizzata (%)'].apply(converti_percentuale)

# Computing video duration in seconds
df['Durata video'] = (df['Durata di visualizzazione media'] / df['Percentuale media visualizzata (%)']) * 100

# Function that converts the video duration from seconds to a HH:mm:ss format
def tempo_in_str(secondi):
    ore = secondi // 3600
    minuti = (secondi % 3600) // 60
    secondi = secondi % 60
    return f"{int(ore):02d}:{int(minuti):02d}:{int(secondi):02d}"

# Applying transformations
df['Durata video'] = df['Durata video'].apply(tempo_in_str)
df['Durata di visualizzazione media'] = df['Durata di visualizzazione media'].apply(tempo_in_str)
```

```python
# Columns renaming
df = df.rename(columns={'Ora pubblicazione video': 'Data pubblicazione'})
df = df.rename(columns={'Commenti aggiunti': 'Commenti'})
df = df.rename(columns={'N. di Non mi piace': 'Non mi piace'})
df = df.rename(columns={'Mi piace: % (%)': 'Mi piace %'})
```

```python
# Computing the number of views per day
df['Data pubblicazione'] = pd.to_datetime(df['Data pubblicazione'])
df['Differenza giorni'] = (datetime.now() - df['Data pubblicazione']).dt.days
df['Visualizzazioni medie al giorno'] = (df['Visualizzazioni'] / df['Differenza giorni']).astype('int')
```

```python
# Filling NaN for non-monetized videos
df['Percentuale di clic delle impressioni (%)'] = df['Percentuale di clic delle impressioni (%)'].str.replace(',', '.').astype(float).round(2)
df['Entrate stimate (EUR)'] = df['Entrate stimate (EUR)'].str.replace(',', '.')
df['RPM (EUR)'] = df['RPM (EUR)'].str.replace(',', '.')
df['CPM (EUR)'] = df['CPM (EUR)'].str.replace(',', '.')
df['Entrate stimate (EUR)'] = df['Entrate stimate (EUR)'].fillna(0)
df['RPM (EUR)'] = df['RPM (EUR)'].fillna(0)
df['CPM (EUR)'] = df['CPM (EUR)'].fillna(0)
df['Entrate stimate (EUR)'] = df['Entrate stimate (EUR)'].astype(float).round(2)
df['RPM (EUR)'] = df['RPM (EUR)'].astype(float).round(2)
df['CPM (EUR)'] = df['CPM (EUR)'].astype(float).round(2)
print(df.isna().sum().sum())
```

```python
df = df[[
    'Titolo video', 
    'Data pubblicazione', 
    'Mi piace %',
    'Mi piace',
    'Non mi piace',
    'Iscritti conquistati',
    'Iscritti persi',
    'Iscritti',
    'Durata video',
    'Percentuale media visualizzata (%)',
    'Durata di visualizzazione media',
    'Visualizzazioni',
    'Visualizzazioni medie al giorno',
    'Impressioni',
    'Percentuale di clic delle impressioni (%)',
    'Commenti',
    'Condivisioni',
    'Entrate stimate (EUR)',
    'RPM (EUR)',
    'CPM (EUR)'
]]
df.head()
```

```python
df.to_csv('youtube_analytics.csv', index=False)
```
